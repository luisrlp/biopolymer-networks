```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
file_name = 'data.out'

with open(file_name, 'r') as file:
    file_values = [float(line.strip()) for line in file]

data = np.array(file_values)
data.shape, data.mean(), data.std(), data.var()
```

```python
plt.hist(data, bins=10)
```
